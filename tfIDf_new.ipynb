```python
import csv
import csv
from nltk import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import operator
```

```python
data = []
count = 0
with open('new_data.csv') as csvfile:
    spamreader=csv.reader(csvfile)
    for row in spamreader:
        info = {}
        #print row[0]
        count += 1
        data.append(row[1])
print "Number of Items == " + str(count)
```

```python
# data
```

```python
vectorizer = TfidfVectorizer(min_df=1,ngram_range=(1, 2))
X = vectorizer.fit_transform(data)
idf = vectorizer.idf_
jobD =  dict(zip(vectorizer.get_feature_names(), idf))
jobD = sorted(jobD.items(), key=operator.itemgetter(1))
```

```python
jobD
```

```python
print "Number of words     == " + str(len(jobD))
# Get top 50 percent values only
percentValue = 0.5
top = int(len(jobD)*percentValue)
top50 =  jobD[-top:]
print "Number of top Words == " + str(len(top50))
```

```python
topWords = []
for t in top50:
    topWords.append(t[0])
topWords = [h.encode('UTF8') for h in topWords]
```

```python
newdata = []
import re
count = 0
for d in data:
    d = d.replace("'"," ")
    d = d.replace("["," ")
    d = d.replace("]"," ")
    d = re.sub('\d', '', d)
    d = re.sub("_", " ", d)
    #print d
    #print len(d)
    d = d.split(" ")
    #print d
    #print d[0]
    d = [w for w in d if w in topWords]
#     print "xxx" + d
    x = " ".join(d)
    if(x != ''):
        count += 1
    newdata.append(x)
```

```python
count = 0
with open("tfOutputs/finalInput.csv","a") as f:
    wr = csv.writer(f,delimiter="\n")
    for d in newdata:
        count += 1
        wr.writerow([d+","])
        print ("-----") + str(count)
```
